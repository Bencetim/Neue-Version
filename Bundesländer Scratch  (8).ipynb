```python
import ipywidgets as widgets
import sys
from IPython.display import display
from IPython.display import clear_output
def create_multipleChoice_widget(description, options, correct_answer):
    if correct_answer not in options:
        options.append(correct_answer)
    
    correct_answer_index = options.index(correct_answer)
    
    radio_options = [(words, i) for i, words in enumerate(options)]
    alternativ = widgets.RadioButtons(
        options = radio_options,
        description = '',
        disabled = False
    )
    
    description_out = widgets.Output()
    with description_out:
        print(description)
        
    feedback_out = widgets.Output()

    def check_selection(b):
        a = int(alternativ.value)
        if a==correct_answer_index:
            s = '\x1b[6;30;42m' + "Super, das ist die richtige Antwort" + '\x1b[0m' +"\n" #green color
        else:
            s = '\x1b[5;30;41m' + "Schade das war leider falsch " + '\x1b[0m' +"\n" #red color
        with feedback_out:
            clear_output()
            print(s)
        return
    
    check = widgets.Button(description="submit")
    check.on_click(check_selection)
    
    
    return widgets.VBox([description_out, alternativ, check, feedback_out])
    

Q1 = create_multipleChoice_widget('Welches Bundesland ist das?',['Saarland','Rheinland-Pfalz','Brandenburg'],'Saarland')
from IPython.display import Image as Saarland, display
display(Saarland('https://www.feuerwehrabzeichen-weltweit.de/Landkreise/Saarland/Karte_Bundesland_Saarland.png', width=300, height=300))
pass
display(Q1)
Q2 = create_multipleChoice_widget('Welches Bundesland ist das?',['Sachsen','Hessen','Berlin'],'Berlin')
from IPython.display import Image as Berlin, display
display(Berlin('https://www.feuerwehrabzeichen-weltweit.de/Landkreise/Berlin/Landkarte_Berlin.png', width=300, height=300))
pass
display(Q2)
Q3 = create_multipleChoice_widget('Welches Bundesland ist das?',['Brandenburg','Bremen','Bayern'],'Brandenburg')
from IPython.display import Image as Brandenburg, display
display(Brandenburg('https://upload.wikimedia.org/wikipedia/commons/thumb/6/6c/Locator_map_Brandenburg_in_Germany.svg/757px-Locator_map_Brandenburg_in_Germany.svg.png', width=300, height=300))
pass
display(Q3)
Q4 = create_multipleChoice_widget('Welches Bundesland ist das?',['Bayern','Hamburg','Hessen'],'Hamburg')
from IPython.display import Image as Hamburg, display
display(Hamburg('https://upload.wikimedia.org/wikipedia/commons/thumb/c/cc/Locator_map_Hamburg_in_Germany.svg/1200px-Locator_map_Hamburg_in_Germany.svg.png', width=300, height=300))
pass
display(Q4)
Q5 = create_multipleChoice_widget('Welches Bundesland ist das?',['Niedersachsen','Baden-Württemberg','Nordrhein-Westfalen'],'Niedersachsen')
from IPython.display import Image as Niedersachsen, display
display(Niedersachsen('https://upload.wikimedia.org/wikipedia/commons/thumb/f/f7/Locator_map_Lower-Saxony_in_Germany.svg/1200px-Locator_map_Lower-Saxony_in_Germany.svg.png', width=300, height=300))
pass
display(Q5)
Q6 = create_multipleChoice_widget('Welches Bundesland ist das?',['Bayern','Schleswig-Holstein','Hessen'],'Hessen')
from IPython.display import Image as Hessen, display
display(Hessen('https://upload.wikimedia.org/wikipedia/commons/thumb/7/77/Locator_map_Hesse_in_Germany.svg/757px-Locator_map_Hesse_in_Germany.svg.png', width=300, height=300))
pass
display(Q6)
Q7 = create_multipleChoice_widget('Welches Bundesland ist das?',['Bremen','Hamburg','Nordrhein-Westfalen'],'Bremen')
from IPython.display import Image as Bremen, display
display(Bremen('https://www.feuerwehrabzeichen-weltweit.de/Landkreise/Bremen/Landkarte_Bremen.png', width=300, height=300))
pass
display(Q7)

```
